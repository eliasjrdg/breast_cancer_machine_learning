## ***Imports***

```python
import os
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('pastel')
sns.set_style("whitegrid")
```

## ***Obtendo os dados***

```python
dataset = pd.read_csv(os.path.join(os.getcwd(), 'breast-cancer-wisconsin-data.csv'))
dataset.head()
```

* Verificando se existem registro vazios

```python
dataset.info()
```

```python
dataset.describe()
```

## ***Data Cleansing***

* Removendo a coluna 'id'

```python
headers = list(dataset.columns.values)
headers = headers[1:]
print(headers)
```

```python
df = dataset[headers].copy()
df.head()
```

* Transformando a coluna 'diagnosis' em 0=Benign e 1=Malignant

```python
df['diagnosis'].value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].replace({'M':1, 'B':0})
```

```python
df['diagnosis'].value_counts()
```

## ***EDA - Exploratory Data Analysis***

```python
df.head()
```

```python
# verificando a relação entre 'área média', 'raio médio', 'textura média' e 'diagnóstico'
fig, axs = plt.subplots(ncols=3, figsize=(15, 5))
sns.histplot(df, x='area_mean', bins=30, multiple='stack', hue='diagnosis', ax=axs[0])
sns.histplot(df, x='radius_mean', bins=30, multiple='stack', hue='diagnosis', ax=axs[1])
sns.histplot(df, x='texture_mean', bins=30, multiple='stack', hue='diagnosis', ax=axs[2])

plt.show()
```

```python
# verificando a relação entre algumas variáveis e o diagnóstico
features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']

sns.pairplot(df, vars=features, hue='diagnosis')

plt.show()
```

## ***Modelo de RandomForestClassifier***

* Separando os dados de teste e treino

```python
from sklearn.model_selection import train_test_split
```

```python
num_var = list(df.columns.values)
num_var = num_var[1:]
print(num_var)
```

```python
X = df[num_var]
Y = df['diagnosis']
```

```python
X.head()
```

```python
Y.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

* Trainando o modelo de RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rnd_for = RandomForestClassifier(random_state=42)
```

```python
rnd_for.fit(x_train, y_train)
```

* Avaliando o modelo

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
rnd_predictions = rnd_for.predict(x_test)
```

```python
print('Accuracy:', (accuracy_score(y_test, rnd_predictions) * 100), '\n')
print(classification_report(y_test, rnd_predictions))
```

```python
cf_matrix = confusion_matrix(y_test, rnd_predictions)

print(cf_matrix)

plt.figure(figsize=(6, 4))
sns.heatmap(cf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

* Verificando a impotância das features

```python
importances = rnd_for.feature_importances_
indices = np.argsort(importances)[::-1]
```

```python
palette = sns.color_palette("Blues", n_colors=x_train.shape[1])[::-1]

plt.figure(figsize=(12, 6))
plt.title('Feature Importances')
sns.barplot(x=x_train.columns[indices], y=importances[indices], palette=palette)
plt.xticks(range(x_train.shape[1]), x_train.columns[indices], rotation=90)
plt.tight_layout()
plt.show()
```
